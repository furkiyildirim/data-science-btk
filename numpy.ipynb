# ARRAY INDEX METHODS

```python
numArr = np.arange(0, 15)
numArr
```

```python
# GET INDEX
numArr[0]
```

```python
# SLICE
numArr[0:5]
```

```python
# EDIT
numArr[0:5] = 3
numArr # values is changed
```

```python
numArrN = np.arange(0, 24)
numArrN
```

```python
mergedArray = numArrN[2:10]
mergedArray
```

```python
mergedArray[:] = 2
mergedArray
```

```python
numArrN # What the fuck moment
# this is about numpy state
```

```python
# to avoid this situation use '.copy()' func
nArr1 = np.arange(0, 24)
nArr2 = nArr1.copy()[2:10]
nArr2[:] = 2
print('nArr1: ', nArr1)
print('nArr2: ', nArr2)
# As a result, original values of nArr1 is not changed thanks to the .copy() method
```
