```python
import pandas as pd # importance
import numpy as np # we usually use together
```

## PANDAS SERIES

```python
dictionary:dict = {
    "Atil": 50,
    "Zeynep": 40,
    "Mehmet": 30,
}
```

```python
pd.Series(dictionary)
```

```python
type(pd.Series(dictionary))
```

```python
array1 = np.array([50, 40, 30])
array2 = np.array(['Atil', 'Zeynep', 'Mehmet'])
# You can also use normal python list for series
```

```python
pd.Series(array2)
```

```python
pd.Series(array1, array2)  # first parr: key , second: value
```

```python
# or
pd.Series(data = array1, index = array2)
```
